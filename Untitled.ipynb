```python
import json
import os
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer
import ast
from sklearn.feature_extraction.text import TfidfVectorizer
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem import PorterStemmer
import string 
from nltk.corpus import stopwords

from vowpalwabbit.DFtoVW import DFtoVW, SimpleLabel, Feature
import pandas as pd
import artm
```

```python
def extract_utterance(data):
    for item in data:
        for turn in item['turns']:

            text = turn['utterance']
            texts.append(text)
def extract_texts(file_list):   
    for file in file_list:
        with open('data/{}'.format(file)) as f:
            data = json.load(f)
            extract_utterance(data)
            
texts  = []
data = 'data'
file_list = os.listdir(data)
extract_texts(file_list)
```

```python
nltk.download("stopwords")
nltk.download('punkt')

stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()
```

```python
def clean(text):
  tokens = word_tokenize(text)
  tokens = [w.lower() for w in tokens]
  table = str.maketrans('', '', string.punctuation)
  stripped = [w.translate(table) for w in tokens]
  words = [word for word in stripped if word.isalpha()]
  words = [w for w in words if not w in stop_words]
  

  stemmed_words = [stemmer.stem(word) for word in words]
  sentence = " ".join(stemmed_words)
  return sentence


cleaned_text = []
for text in texts:
  cleaned_text.append(clean(text))
```

```python
tfidf = TfidfVectorizer()

response = tfidf.fit_transform(cleaned_text)

print(len(tfidf.vocabulary_))

df1 = pd.DataFrame(response.toarray(), columns=tfidf.get_feature_names())

df1
```

```python
from sklearn.metrics.pairwise import linear_kernel
cosine_similarities = linear_kernel(df1)
```

```python
print(len(cosine_similarities))
cosine_similarities
```

```python
#df = pd.DataFrame({"y": [1], "a": [2], "b": [3], "c": [4]})
conv1 = DFtoVW(df=df1,
                    features=[Feature(col) for col in df1.columns])

vw = conv1.convert_df()

type(vw)
```

```python
with open('vwdata.txt', 'w') as vwdata:
  for v in vw:
    vwdata.write(v)
```
